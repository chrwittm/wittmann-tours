# Creating a `zip`-file containing all blog posts

The goal of this notebook is to provide a zip file containing all blog posts.

```python
!rm ./zip/blogposts.zip
!rm ./zip/blogposts-md.zip
```

```python
# all files
!(cd blogposts && zip -r ../zip/blogposts.zip .)
```

```python
# only .jpg and .md files
!(cd blogposts && find . -type f \( -iname "*.jpg" -o -iname "*.md" \) -exec zip -r ../zip/blogposts-md.zip {} +)
```

The command above works as follows:
- `(cd blogposts && ...)` changes the directory to `blogposts` for the execution of the command inside the parentheses, ensuring that the rest of the command operates within this directory.
- `find . -type f` starts the search from the current directory (`.`), looking for files (`-type f`).
- `\( -iname "*.jpg" -o -iname "*.md" \)` specifies the search criteria: files that end with `.jpg` or `.md`, ignoring case (`-iname` for case-insensitive name matching).
- `-exec zip -r ../zip/blogposts-md.zip {} +` adds the found files to the ZIP archive. `{}` is replaced by the found file paths, and `+` at the end tells `find` to add all found files to the zip command at once, where possible, for efficiency.
